**Identifying and Removing Outliers**

```python
df.describe()
```

```python
sns.boxplot(df.adr)
```

```python
Q1 = df['adr'].quantile(0.25)
Q3 = df['adr'].quantile(0.75)
IQR = Q3-Q1

print('Q1:', Q1)
print('Q3:', Q3)
print('IQR:', IQR)
```

```python
max_IQR = Q3 + 1.5 * IQR
min_IQR = Q1 - 1.5 * IQR

print('min_IQR:', min_IQR)
print('max_IQR:', max_IQR)
```

```python
outliers = df.loc[(df['adr'] > max_IQR) | (df['adr'] < min_IQR)]
```

```python
new_df = len(df['adr']) - len(outliers)
print(new_df)
```

```python
new_df = df.loc[(df['adr'] <= max_IQR) & (df['adr'] >= min_IQR)]
print(new_df)
```

```python
sns.boxplot(new_df.adr)
```
